```python
import pandas as pd

# CSV 파일을 불러오는 예시
file_path = 'Meteorite_Landings.csv'

# CSV 파일을 데이터프레임으로 읽어오기
df = pd.read_csv(file_path)

# 데이터프레임 확인
df.head() # 처음 5개의 행 출력
```

```python
# 모든 null 값을 삭제
df.dropna(inplace=True)

# 삭제된 데이터프레임 확인
df
```

```python
df.shape
```

```python
df.describe()
```

```python
# 데이터프레임에서 year 컬럼의 최댓값이 2023보다 큰 행을 제거
df = df[df['year'] <= 2023]
```

```python
df.describe()
```

```python
from matplotlib import font_manager, rc

# 한글 폰트 설정 (Windows 환경에서는 다른 폰트를 사용할 수 있음)
font_path = "C:/Windows/Fonts/malgun.ttf"  # 사용할 한글 폰트 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# 'mass (g)' 컬럼 이름을 'mass'로 변경
df.rename(columns={'mass (g)': 'mass'}, inplace=True)
```

```python
df.head()
```

```python
# 데이터프레임에서 위도와 경도가 0.0인 행을 제거
df = df[(df['reclat'] != 0.0) & (df['reclong'] != 0.0)]
```

```python
df.shape
```

```python
import geopandas as gpd
from shapely.geometry import Point

# 위도와 경도 값을 활용하여 GeoDataFrame 생성
geometry = [Point(xy) for xy in zip(df['reclong'], df['reclat'])]
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# 지리 정보 데이터셋 로드 (이 데이터셋은 필요에 따라 다르게 설정해야 합니다)
# 예시: 대륙별 경계 정보를 담은 shapefile 사용
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# 좌표를 지역으로 변환
gdf['region'] = None  # 빈 'region' 열 추가

for index, row in gdf.iterrows():
    point = row['geometry']
    for index2, row2 in world.iterrows():
        if point.within(row2['geometry']):
            gdf.at[index, 'region'] = row2['name']

# 결과 출력
print(gdf[['name', 'reclat', 'reclong', 'region']])
```

```python
# CSV 파일로 저장
gdf.to_csv('Meteorite_Landings_changed_data.csv', index=False, encoding='utf-8-sig')
```
