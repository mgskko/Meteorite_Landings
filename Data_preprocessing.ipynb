```python
import pandas as pd

# CSV 파일을 불러오는 예시
file_path = 'Meteorite_Landings.csv'

# CSV 파일을 데이터프레임으로 읽어오기
df = pd.read_csv(file_path)

# 데이터프레임 확인
df.head() # 처음 5개의 행 출력
```

```python
# 모든 null 값을 삭제
df.dropna(inplace=True)

# 삭제된 데이터프레임 확인
df
```

```python
from matplotlib import font_manager, rc

# 한글 폰트 설정 (Windows 환경에서는 다른 폰트를 사용할 수 있음)
font_path = "C:/Windows/Fonts/malgun.ttf"  # 사용할 한글 폰트 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# 'mass (g)' 컬럼 이름을 'mass'로 변경
df.rename(columns={'mass (g)': 'mass'}, inplace=True)

# CSV 파일로 저장
df.to_csv('Meteorite_Landings_changed_data.csv', index=False, encoding='utf-8-sig')
```
